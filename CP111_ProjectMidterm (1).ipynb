````python
''' 
รายชื่อกลุ่ม
    นิธิศ เกิดช่วย              
    64102010078
    นายปุณยเกียรติ ถนอมทรัพย์ 
    64102010080
'''
import re
# Funtion text to sentence using split


def sentence(text):
    newstr = []
    ls = list(text.split('.'))
    for sub in ls:
        newstr.append(sub.replace("\n", ""))
    print(newstr)

# Funtion text to non duplicate


def non_duplicate(text):
    text = text.replace("it's","it is")
    words_str = text.split(" ")
    result = []
    for word in words_str:
        # store unique words in list
        s = word.replace(".", "")
        if (text.count(word) >= 1 and (word.lower() not in result) and (word not in result) and (s not in result)):
            result.append(word)
    print(" ".join(result))

# Funtion text to word frequency using count() and then store a count to dictionary


def wordfreq(text):
    text = text.replace("it's","it is")
    wordList = re.sub("[^\w]", " ", text).lower().split()
    word_freq = {
        "this": 0, "these": 0, "those": 0, "a": 0, "an": 0, "the": 0, "is": 0, "am": 0, "are": 0
    }
    for i in word_freq.keys():
        word_freq.update({i: wordList.count(i)})
    print(word_freq)


# Main Program
welcome = '''
                      ⣿⣷⡶⠚⠉⢀⣤⣾⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡿⠋⠠⣴⣿⣿⣿⣿⣿⣿⣿
                      ⠿⠥⢶⡏⣸⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡿⠋⢀⣴⣷⣌⢿⣿⣿⣿⣿⣿⣿
                      ⣍⡛⢷⣠⣿⣿⣿⣿⣿⣟⠻⣯⠽⣿⣿⠟⠁⣠⠿⠿⣿⣿⣎⠻⣿⣿⣿⣿⣿
                      ⣿⣿⣦⠙⣿⣿⣿⣿⣿⣿⣷⣏⡧⠙⠁⣀⢾⣧   ⠈⣿⡟⠙⣫⣵⣶⣿⣿
                      ⣿⣿⣿⢀⣿⣿⣿⣿⣿⣿⣿⠟⠃⢀⣀⢻⣎⢻⣷⣤⣴⠟  ⣠⣾⣿⢟⣵⢿
                      ⣿⣯⣄⢘⢻⣿⣿⣿⣿⡟⠁⢀⣤⡙⢿⣴⣿⣷⡉⠉⢀  ⣴⣿⡿⣡⣿⣿⡿
                      ⠿⣿⣧⣤⡘⢿⣿⣿⠏  ⡔⠉⠉⢻⣦⠻⣿⣿⣶⣾⡟⣼⣿⣿⣱⣿⡿⢫⣾
                      ⣷⣮⣝⣛⣃⡉⣿⡏  ⣾⣧⡀  ⣿⡇⢘⣿⠋    ⠻⣿⣿⣿⢟⣵⣿⣿
                      ⣿⣿⣿⣿⣿⣿⣌⢧⣴⣘⢿⣿⣶⣾⡿⠁⢠⠿⠁⠜    ⣿⣿⣿⣿⡿⣿⣿
                      ⣿⣿⣿⣿⣿⣿⣿⣦⡙⣿⣷⣉⡛⠋    ⣰⣾⣦⣤⣤⣤⣿⢿⠟⢋⣴⣿⣿
                      ⣿⣿⣿⣿⣿⣿⣿⣿⣿⣌⢿⣿⣿⣿⣿⢰⡿⣻⣿⣿⣿⣿⣿⢃⣰⣫⣾⣿⣿⣿
                      ⢿⣿⣿⣿⣿⣿⣿⣿⣿⣿⡆⠿⠿⠿⠛⢰⣾⡿⢟⣭⣿⣿⣿⣿⣿⣿⣿⣿⣿⣿

                  Welcome to text program of midteam project CP111

                        > 1.  Text to sentence list
                        > 2.  Text to non duplicate
                        > 3.  Text to word frequency

                        > 0.  Exit program
    '''
print(welcome)
while True:
    action = input("Please select program or Exit : ")
    if action == "0":
        print("Good bye!")
        break
    #Enter text and go to the function 
    text = input("Enter your text : ")
    if action == "1":
        sentence(text)
    elif action == "2":
        non_duplicate(text)
    elif action == "3":
        wordfreq(text)
    else:
        print("Enter again!")
        continue
````
